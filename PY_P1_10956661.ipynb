###  Replacing null values with the mean of each column

```python
for column in data.columns:
    if column != 'ID':
       data[column].fillna(data[column].mean(), inplace=True)
```

### **METHOD**: Replacement of null values with the mean in each column.


### Reason for using the method of replacement.
___

When performing an analysis on a dataset, we want to make sure that we have as much information as possible to perform accurate analysis. So when missing values are detected in our data, the whole row must not be deleted. If we simply delete the rows with missing values, we may lose important information that could be useful for our analysis.

Instead, the best strategy is to replace the missing values with a reasonable estimate. One way to do this is to replace the missing values with the mean value of the column. This strategy assumes that the missing values are missing at random and that the mean value is representative of the overall distribution of the data.

By replacing the missing values with the mean value, we can preserve the integrity of the dataset and reduce the impact of missing values on our analysis.This will help prevent inaccuracy too.

### Values above and below the mean value for each column
---

```python
nrows = 3
ncols = 4

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15, 10))

for i, column in enumerate(data.columns):
    if column not in ['ID', 'quality']:
        # Calculating mean of each column
        column_mean = data[column].mean()

        # Creating a boolean series based on condition
        compared_column_mean = data[column] < column_mean

        # Defining labels and explode values
        labels = ['Above Mean', 'Below Mean']
        explode = [0.1, 0]

        # Creating pie chart for each column
        sns.set_palette('rocket')
        axes[i//ncols, i%ncols].pie(compared_column_mean.value_counts(), explode=explode, labels=labels, shadow=True)
        axes[i//ncols, i%ncols].set_title(f'{column} values ')

plt.tight_layout()


plt.show()
```

### Z_SCORES AND OUTLIERS
**z-score** represents the number of standard deviations that a data point is away from the mean of the distribution.

**outlier** is a data point that falls significantly outside the range of other data points. One common method for identifying outliers is using a threshold of 3 standard deviations.

```python
for column in data.columns:
    if column != 'ID' :
    #Calculating the z-scores for each data point of each column
       z_scores = (data[column] - data[column].mean()) / data[column].std()
    #output below is all z_scores values for each column
       print(z_scores)

     #outliers are identified using a threshold of 3 standard deviations
       outliers = data[abs(z_scores) > 3]
```

```python
outliers
```

```python
for column in outliers.columns:
    if column != 'ID':
        plt.title(f'Outliers in {column}')
        plt.xlabel('Range')
        
        sns.scatterplot(outliers[column])
        plt.show()
```

### Ranking each columns mean by quality
---
Data is grouped categorically by quality and the mean of each column is computed and ranked.

Groups are sorted in descending order based on the value of the  mean.

```python
for column in data.columns:
    if column != 'ID' :

       grouped_data = data.groupby('quality')[column].mean()


       grouped_data = grouped_data.sort_values(ascending=False)

       print(grouped_data)
       print( '\n\n')
```
